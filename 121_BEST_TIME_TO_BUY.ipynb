### Solution

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        max_low=float('inf')
        max_profit=0
        

        for i in prices:
            if i<max_low:
                max_low = i

            profit=i-max_low

            if profit>max_profit:
                max_profit=profit

        return max_profit
```
