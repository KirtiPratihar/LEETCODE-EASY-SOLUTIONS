### Solution

```python

class Solution:
    def isHappy(self, n: int) -> bool:
        def next_num(numb):
            sum=0
            while numb > 0:
                digit=numb % 10
                sum=sum+(digit**2)
                numb = numb // 10
            return sum
        done=set()
        while n!=1 and n not in done:
            done.add(n)
            n=next_num(n)
        return n==1

        
```
